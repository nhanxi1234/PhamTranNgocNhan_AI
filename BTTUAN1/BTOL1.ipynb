```python
#bai11

# Get the number of elements for array a
try:
    n = int(input("Nhập số lượng phần tử cho mảng a (n): "))
except ValueError:
    print("Giá trị nhập vào không hợp lệ. Vui lòng nhập một số nguyên.")
    n = 0 # Default to 0 elements if input is invalid

# Create array a with n integer elements
a = []
if n > 0:
    print(f"Nhập {n} phần tử cho mảng a:")
    for i in range(n):
        while True:
            try:
                element = int(input(f"Phần tử thứ {i+1}: "))
                a.append(element)
                break
            except ValueError:
                print("Giá trị nhập vào không hợp lệ. Vui lòng nhập một số nguyên.")

# Create array b containing only even elements from a
b = []
for element in a:
    if element % 2 == 0:
        b.append(element)

print(f"Mảng a: {a}")
print(f"Mảng b (chỉ chứa các phần tử chẵn của a): {b}")
```

```python
#bai13
def merge_arrays_sum_and_append(arr1, arr2):

    result = []
    len1 = len(arr1)
    len2 = len(arr2)
    min_len = min(len1, len2)

    # Calculate sum for the common part
    for i in range(min_len):
        result.append(arr1[i] + arr2[i])

    # Append remaining elements from the longer array, if any
    if len1 > len2:
        result.extend(arr1[min_len:])
    elif len2 > len1:
        result.extend(arr2[min_len:])

    return result

# Example usage based on the problem description
a = [3, 9, 1, 4]
b = [2, 7, 4, 3, 2, 8]
result_array = merge_arrays_sum_and_append(a, b)
print(f"Mảng a: {a}")
print(f"Mảng b: {b}")
print(f"Mảng kết quả: {result_array}")
```
