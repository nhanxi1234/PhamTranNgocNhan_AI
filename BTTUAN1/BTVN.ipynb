```python
#bai1
import numpy as np
def tron_mang(a,b):
  c =[]
  i = 0
  while i < len(a) and i < len(b):
    c.append(min(a[i],b[i]))
    i += 1
  if i<len(a):
    c.extend(a[i:])
  if i<len(b):
    c.extend(b[i:])
  return c
a = [3,9,1,4]
b = [2,7,4,3,2,8]
print(tron_mang(a,b))
```

```python
#bai2
def tong_tam_gia_tren_duong_cheo_phu(a,n):
  s = 0;
  for i in range(n):
    for j in range(n):
      if i+j <= n-1:
        s += a[i][j]
  return s;
def chuyen_phan_tu_am(a):
  for i in range(len(a)):
    for j in range(len(a[0])):
      if (a[i][j]<0):
        a[i][j] = abs(a[i][j])
  return a
def thay_chan(a,x):
  for i in range(len(a)):
    for j in range(len(a[0])):
      if a[i][j] % 2 ==0:
        a[i][j] = x
  return a
def kiem_tra_toan_chan(a):
  for row in a:
    for val in row:
      if (val % 2 != 0):
        return False
  return True
def doi_xung(a,n):
  for i in range(n):
    for j in range(n):
      if a[i][j] != a[j][i]:
        return False
  return True
def duong_cheo_chinh_tang_dan(a,n):
  for i in range(n-1):
    if (a[i][i] > a[i+1][i+1]):
      return False
  return True
def xuat_tam_giac_duoi_duong_cheo_phu(a,n):
  for i in range(n):
    for j in range(n):
      if i+j >= n-1:
        print(a[i][j],end = " ")
    print()
def kiem_tra_ma_tran_cheo_phu_giam_dan(a,n):
  for i in range(n - 1):
    if a[i][n-1-i] < a[i+1][n-2-i]:
      return False
  return True

a = [
    [1,2,-3],
    [4,5,6],
    [-7,8,9]
]
n=3
print("Tong tam giac duong cheo phu: ",tong_tam_gia_tren_duong_cheo_phu(a,n))
print("Mang sau khi chuyen phan tu sang duong: ")
print(chuyen_phan_tu_am(a))
print ("thay phan tu chan sang 0: ")
print(thay_chan(a,0))
print("Toan chan: ",kiem_tra_toan_chan(a))
print("Doi xung: ",doi_xung(a,n))
print("Cheo chinh tang:", duong_cheo_chinh_tang_dan(a,n))
print("Tam giac duoi cheo phu: ")
xuat_tam_giac_duoi_duong_cheo_phu(a,n)
print("Cheo phu giam:",kiem_tra_ma_tran_cheo_phu_giam_dan(a,n))
```
