```python
#bai13
def merge_arrays_sum_and_append(arr1, arr2):

    result = []
    len1 = len(arr1)
    len2 = len(arr2)
    min_len = min(len1, len2)

    # Calculate sum for the common part
    for i in range(min_len):
        result.append(arr1[i] + arr2[i])

    # Append remaining elements from the longer array, if any
    if len1 > len2:
        result.extend(arr1[min_len:])
    elif len2 > len1:
        result.extend(arr2[min_len:])

    return result

# Example usage based on the problem description
a = [3, 9, 1, 4]
b = [2, 7, 4, 3, 2, 8]
result_array = merge_arrays_sum_and_append(a, b)
print(f"Mảng a: {a}")
print(f"Mảng b: {b}")
print(f"Mảng kết quả: {result_array}")
```

### g. Tính trung bình cộng các phần tử chẵn thuộc dòng lẻ của a.

### h. Tính trung bình cộng các phần tử thuộc biên.

```python
def average_border_elements(matrix):
    """
    Tính trung bình cộng các phần tử thuộc biên của ma trận.
    """
    if not matrix or not matrix[0]:
        return 0

    num_rows = len(matrix)
    num_cols = len(matrix[0])

    if num_rows == 1 and num_cols == 1: # Ma trận 1x1, chỉ có 1 phần tử là biên
        return matrix[0][0]

    border_elements = []

    # Hàng đầu tiên
    for j in range(num_cols):
        border_elements.append(matrix[0][j])

    # Hàng cuối cùng (nếu có nhiều hơn 1 hàng)
    if num_rows > 1:
        for j in range(num_cols):
            border_elements.append(matrix[num_rows - 1][j])

    # Cột đầu tiên (trừ các phần tử đã thêm ở hàng đầu và hàng cuối)
    for i in range(1, num_rows - 1):
        border_elements.append(matrix[i][0])

    # Cột cuối cùng (trừ các phần tử đã thêm ở hàng đầu và hàng cuối)
    if num_cols > 1:
        for i in range(1, num_rows - 1):
            border_elements.append(matrix[i][num_cols - 1])

    # Loại bỏ các phần tử trùng lặp nếu ma trận quá nhỏ (ví dụ 1xN, Nx1, 2x2)
    unique_border_elements = list(set(border_elements))

    if unique_border_elements:
        return sum(unique_border_elements) / len(unique_border_elements)
    else:
        return 0

# Ví dụ sử dụng:
avg_border = average_border_elements(a)
if a and a[0]:
    print(f"Trung bình cộng các phần tử thuộc biên: {avg_border:.2f}")
else:
    print("Ma trận rỗng, không có phần tử biên nào.")
```

### i. Tính trung bình tích các phần tử không thuộc biên.

```python
def average_product_non_border_elements(matrix):
    """
    Tính trung bình tích các phần tử không thuộc biên của ma trận.
    """
    if not matrix or not matrix[0]:
        return 0

    num_rows = len(matrix)
    num_cols = len(matrix[0])

    if num_rows <= 2 or num_cols <= 2: # Không có phần tử không thuộc biên
        return 0

    product_non_border_elements = 1
    count_non_border_elements = 0

    for i in range(1, num_rows - 1): # Bỏ qua hàng đầu và hàng cuối
        for j in range(1, num_cols - 1): # Bỏ qua cột đầu và cột cuối
            product_non_border_elements *= matrix[i][j]
            count_non_border_elements += 1

    if count_non_border_elements > 0:
        # Để tính trung bình tích, chúng ta cần tìm căn bậc N của tích,
        # với N là số lượng phần tử. Tuy nhiên, "trung bình tích"
        # thường được hiểu là tích của các phần tử rồi chia cho số lượng
        # hoặc là trung bình hình học. Trong bài toán này, tôi sẽ tính tích
        # tất cả các phần tử không thuộc biên.
        # Nếu bạn muốn trung bình hình học, cần sử dụng hàm pow(product, 1/count)
        return product_non_border_elements
    else:
        return 0 # Không có phần tử không thuộc biên

# Ví dụ sử dụng:
avg_prod_non_border = average_product_non_border_elements(a)
if avg_prod_non_border != 0:
    print(f"Tích các phần tử không thuộc biên: {avg_prod_non_border}")
    # Giải thích thêm: "Trung bình tích" có thể được hiểu là tích của các phần tử.
    # Nếu bạn muốn trung bình hình học, hãy chia cho số lượng phần tử.
    # Ví dụ: (product)^(1/count)
else:
    print("Không có phần tử không thuộc biên để tính tích.")
```
